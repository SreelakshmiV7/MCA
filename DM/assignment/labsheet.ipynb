```python
import pandas as pd 
import numpy as np
```

```python
# 1Ô∏è‚É£ Load the dataset into a DataFrame and print it
df = pd.read_csv("superstore.csv")
print("=== Entire Dataset ===")
print(df)
```

```python
# 2Ô∏è‚É£ Display the size of the dataset
print("\n=== Size of Dataset ===")
print(df.shape)   # (rows, columns)
```

```python
# 3Ô∏è‚É£ Print the first 10 rows
print("\n=== First 10 Rows ===")
print(df.head(10))
```

```python
# 4Ô∏è‚É£ Print the last 10 rows
print("\n=== Last 10 Rows ===")
print(df.tail(10))
```

```python
# 5Ô∏è‚É£ Display the information about data in the dataset
print("\n=== Dataset Info ===")
print(df.info())
```

```python
# 6Ô∏è‚É£ Display the statistical view of the dataset
print("\n=== Statistical Summary ===")
print(df.describe(include='all'))
```

```python
# 7Ô∏è‚É£ Display the datatypes of data in the dataset
print("\n=== Data Types ===")
print(df.dtypes)
```

```python
# 8Ô∏è‚É£ Print the null values of all columns
print("\n=== Null Values in Each Column ===")
print(df.isnull().sum())
```

```python
# 9Ô∏è‚É£ Display the number of unique values in the dataset
print("\n=== Unique Values in Each Column ===")
print(df.nunique())
```

```python
# üîü Print % of null value in each column
print("\n=== Percentage of Null Values in Each Column ===")
print((df.isnull().sum() / len(df)) * 100)
```

```python
# 1Ô∏è‚É£1Ô∏è‚É£ Fill missing values in the dataset with mean of that column
print("\n=== Filling Missing Values with Column Mean ===")
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        # df[col].fillna(df[col].mean(), inplace=True)  will get deprecated in future...
        df.fillna({col: df[col].mean()}, inplace=True) # Alternative method
print(df.isnull().sum())
```

```python
# 1Ô∏è‚É£2Ô∏è‚É£ Sort the values based on the third column
third_col = df.columns[2]
print(f"\n=== Sorting by Third Column ({third_col}) ===")
df_sorted = df.sort_values(by=third_col)
print(df_sorted)
```

```python
# 1Ô∏è‚É£3Ô∏è‚É£ Rename the second column name
second_col = df.columns[1]
new_col_name = second_col + "_new"
df.rename(columns={second_col: new_col_name}, inplace=True)
print(f"\n=== Renamed '{second_col}' to '{new_col_name}' ===")
print(df.head())
```

```python
# 1Ô∏è‚É£4Ô∏è‚É£ Print the categorical and numerical columns
print("\n=== Categorical Columns ===")
cat_cols = df.select_dtypes(include=['object']).columns
print(cat_cols)

print("\n=== Numerical Columns ===")
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(num_cols)
```

```python
# 1Ô∏è‚É£5Ô∏è‚É£ Drop the outliers from numerical columns using IQR
print("\n=== Removing Outliers ===")
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1
df_no_outliers = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Original shape:", df.shape)
print("After removing outliers:", df_no_outliers.shape)
print("\n=== Dataset after removing outliers ===")
print(df_no_outliers.head())
```
