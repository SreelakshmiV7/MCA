```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
regions = ['North', 'South', 'East', 'West']
products = ['Laptop', 'Smartphone', 'Tablet', 'Headphones', 'Camera']

data = {
    'Month': np.random.choice(months, 200),
    'Region': np.random.choice(regions, 200),
    'Product': np.random.choice(products, 200),
    'Units_Sold': np.random.randint(10, 500, 200),
    'Revenue': np.random.randint(1000, 100000, 200),
    'Discount(%)': np.random.randint(0, 30, 200)
}

df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)
print("✅ File saved as sales_data.csv")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import
df = pd.read_csv('sales_data.csv')

# Step 2: Inspect
print(df.head())       # first 5 rows
print(df.info())       # column info

# Step 3: Clean
df.drop_duplicates(inplace=True)
df.fillna(0, inplace=True)

# Step 4: Analyze
monthly_sales = df.groupby('Month')['Revenue'].sum()

# Step 5: Visualize
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```
